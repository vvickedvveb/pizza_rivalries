```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

class MyStock():
    def __init__(self):
        pass

    
    def create_DF(self, cols, inc_cols, file_name):
        """ Create the company & quotes DataFrame """
        
        def __custom_inc_cols(temp_inc_cols):
            """ Extract columns from 'stock_quotes.csv' """
            temp_cols_to_include = str(temp_inc_cols[2:7]) + str(temp_inc_cols[8:len(temp_inc_cols):3])
            temp_cols_to_include = temp_cols_to_include.replace('[', '').replace(']', '').replace("\'", '')
            temp_cols_to_include = temp_cols_to_include[0:28] + ', ' + temp_cols_to_include[28:]
            temp_cols_to_include = list(temp_cols_to_include.split(', '))
            
            return temp_cols_to_include
         
        
        new_df = pd.read_csv(file_name, names=cols)
        
        if inc_cols == 'custom':
            inc_cols = __custom_inc_cols(cols)

        return new_df[inc_cols]
    

    def df_by_year(self, df, year, ticker):
        """ Create new DF by year, ticker. """
        return df.loc[df['Year'] == year].loc[df['Ticker'] == ticker]


    def dfs_to_plt(self, dfs, x_col, y_col):
        """ Plot continuous line of price by year. """
        for df in dfs:
            #
            xpoints = df[x_col]
            ypoints = df[y_col]
            plt.plot(xpoints, ypoints, label=max(df.Year))
     

    def hist_to_plot(self, dfs, df_column, bin_size = 20, fig_size=(20, 10)):
        """Displays histogram (continuous bar chart) plot based on tuple of dfs by date.

        Parameters
        ----------
        dfs_to_plot : tuple
            A tuple of dataframes usually by date range
            
        df_column : str
            The column series of the DF to plot
            
        bin_size    : int
            How many towers/bins to display (default: 20)
            
        fig_size    : tuple
            Width & height of histogram (default: (20, 10))

        Returns
        -------
            None
        """
        list_years = [] # List of years for bar chart

        plt.figure(figsize=fig_size)

        for df_to_plot in dfs:
            list_years.append(df_to_plot['Year'].unique())
            plt.hist(df_to_plot[df_column], bins=bin_size);

        plt.legend(list_years);
        return


    def get_mean_values_by_year(self, dfs_to_plot = ()):
        """Returns mean value of a column, of a tuple of dataframes.

        Parameters
        ----------
        dfs_to_plot : tuple
            A tuple of dataframes usually by date range

        Returns
        -------
        mean_close_values : list
            A list of the mean value of a certain colummn
            
        pie_label : list
            The year with mean value
        """
        mean_close_values = []
        pie_label = []
        explode_vals = []

        for df_to_plot in dfs_to_plot:
            mean_close_values.append( int(df_to_plot['Close'].mean()) )
            pie_label.append( str(df_to_plot.Year.max()) + ': ' + str(int(df_to_plot['Close'].mean()) ))
            explode_vals.append( df_to_plot['Close'].mean())
            
        return mean_close_values, pie_label, explode_vals


    def get_full_names(self):
        """ Not used  """
        pass
        #return self.pizza_cos.keys()
        
        #for co_name in self.pizza_cos.keys():
        #    print(co_name)
        #    if ' ' in co_name:
        #        temp_co_name = co_name[0].replace('_', ' ')[0:-1].title() + '\'s'
        #list(dict_co_names)
        #papa_name = co_name[0].replace('_', ' ')[0:-1].title() + '\'s'
        #dominos_name = co_name[1][0:-1].title() + '\'s'
        #return (papa_name, dominos_name)
```
