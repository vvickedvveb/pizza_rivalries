```python
def produce_summary_table(df):
    column_renaming = {'count': 'N', 'mean': 'Mean', 'std': 'Std. Dev.', 'min': 'Min.', 'max': 'Max.'}
    column_order = ['Mean', 'Std. Dev.', 'Min.', 'Max.', 'N']

    df_summary = df.describe().rename(column_renaming).T.drop(['25%', '50%', '75%'], axis=1)[column_order]
    df_summary['N'] = df_summary['N'].astype(int)
    return df_summary
    
df_summary = produce_summary_table(df)
df_summary
```
